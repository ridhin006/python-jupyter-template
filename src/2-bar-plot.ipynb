### Exercise 2: Bar Plot with Colors

```python
import matplotlib.pyplot as plt
categories = ['A', 'B', 'C', 'D'] 
values = [5, 7, 3, 9]
plt.bar(categories,values,color=['red','green','blue','yellow'])
plt.title('Bar Chart')
plt.xlabel("categories")
plt.ylabel('Values')
plt.show()
```
